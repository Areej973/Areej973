```python
#test1

import os
import re
import json
from google.colab import userdata
from openai import OpenAI
from tavily import TavilyClient
from pypdf import PdfReader
from pydantic import BaseModel, Field

# 1. Fetch and set API keys securely from Colab Secrets
if "OPENAI_API_KEY" not in os.environ:
    os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

if "TAVILY_API_KEY" not in os.environ:
    os.environ["TAVILY_API_KEY"] = userdata.get('TAVILY_API_KEY')

# 2. Initialize OpenAI and Tavily clients
openai_client = OpenAI()
tavily_client = TavilyClient()

# =====================================================================
# PYDANTIC DATA MODELS FOR VALIDATION & STRUCTURED OUTPUTS
# =====================================================================
class CompanyProfile(BaseModel):
    """Defines structural and type constraints for incoming corporate profile data."""
    company_name: str = Field(..., min_length=2, max_length=100, description="The legal or market name of the firm.")
    products_and_prices: str = Field(..., min_length=10, description="Detailed text listing the solutions and pricing models.")
    brand_voice: str = Field(default="Professional", description="The communication tone alignment.")

# =====================================================================
# SECURITY & VALIDATION FUNCTIONS (Input Validation & Prompt Injection Guard)
# =====================================================================
def validate_input_text(text: str) -> bool:
    """Validates user input against basic prompt injection patterns and malicious intent."""
    if not text.strip():
        return False

    malicious_keywords = [
        "ignore previous instructions", "ignore all instructions", "override system",
        "system prompt", "you must forget", "act as a sudo", "bypass restrictions",
        "drop table", "select * from", "delete from"
    ]

    lowered_text = text.lower()
    for keyword in malicious_keywords:
        if keyword in lowered_text:
            print(f"\n[SECURITY ERROR]: Malicious keyword detected: '{keyword}'")
            return False

    if len(text) > 4000:
        print("\n[VALIDATION ERROR]: Input text size exceeds the allowed limit.")
        return False

    return True

def extract_text_from_pdf(pdf_path: str) -> str:
    """Safely reads and extracts text content from an uploaded PDF file."""
    try:
        reader = PdfReader(pdf_path)
        extracted_text = ""
        for page in reader.pages:
            text = page.extract_text()
            if text:
                extracted_text += text + "\n"
        return extracted_text
    except Exception as e:
        print(f"\n[VALIDATION ERROR]: Failed to process the PDF file. Error: {str(e)}")
        return ""

# =====================================================================
# DATA INGESTION & PARSING LAYER
# =====================================================================
print("================ COMPANY PROFILE ACQUISITION ================\n")
input_method = input("Do you want to enter data manually or upload a PDF? (type 'manual' or 'pdf'): ").strip().lower()

current_profile = None

if input_method == "pdf":
    pdf_file_name = input("Enter the PDF file name uploaded in Colab (e.g., profile.pdf): ").strip()
    if not os.path.exists(pdf_file_name):
        raise ValueError("[ERROR]: File not found. Please upload the PDF to the Colab sidebar first.")

    raw_pdf_content = extract_text_from_pdf(pdf_file_name)

    if not validate_input_text(raw_pdf_content):
        raise ValueError("[SECURITY EXCEPTION]: PDF content violated system security guidelines.")

    print("\nProcessing profile details via OpenAI Structured Outputs (Pydantic parsing)...")

    parse_response = openai_client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Extract corporate profiling structures dynamically from unstructured raw documents."},
            {"role": "user", "content": f"Parse the following text content:\n{raw_pdf_content}"}
        ],
        response_format=CompanyProfile,
        temperature=0.1
    )
    current_profile = parse_response.choices[0].message.parsed

elif input_method == "manual":
    comp_name = input("Enter Company Name: ").strip()
    prod_price = input("Enter Products/Solutions Description with Pricing: ").strip()
    voice = input("Enter Brand Voice (e.g., Bold, Corporate, Friendly): ").strip()

    if not (validate_input_text(comp_name) and validate_input_text(prod_price) and validate_input_text(voice)):
        raise ValueError("[SECURITY EXCEPTION]: Input execution blocked due to invalid entries.")

    current_profile = CompanyProfile(company_name=comp_name, products_and_prices=prod_price, brand_voice=voice)
else:
    raise ValueError("[VALIDATION ERROR]: Invalid choice. You must pick 'manual' or 'pdf'.")

company_name = current_profile.company_name
products_and_prices = current_profile.products_and_prices
brand_voice = current_profile.brand_voice

# =====================================================================
# HUMAN IN THE LOOP (HITL) GATEWAY
# =====================================================================
print("\n================ HUMAN-IN-THE-LOOP VERIFICATION ================")
print(f"Company Name: {company_name}")
print(f"Products & Pricing Structure: {products_and_prices}")
print(f"Brand Voice Alignment: {brand_voice}")
print("=================================================================")

human_approval = input("Review the validated profile data above. Do you approve it? (type 'yes' to proceed, 'no' to abort): ").strip().lower()

if human_approval != 'yes':
    print("\n[SYSTEM TERMINATION]: Execution stopped by user during the validation loop.")
    raise SystemExit

print("\n--- Human Verification Passed. Proceeding to execution phase. ---")

# =====================================================================
# RETRIEVAL: Corporate Intelligence Acquisition via Tavily API (UPDATED LIMIT)
# =====================================================================
# Increased cut-off limit: Trim products description to 250 characters for richer context
clean_products_summary = products_and_prices[:250] if len(products_and_prices) > 250 else products_and_prices

search_query = f"top competitor companies competing with {company_name} in {clean_products_summary}"

# Safety net fallback to guarantee keeping total characters below Tavily's 400 boundary
if len(search_query) > 380:
    search_query = f"top competitor firms competing with {company_name}"

print(f"\n--- Tavily is searching live web using safe query ({len(search_query)} chars): '{search_query}' ---")
search_result = tavily_client.search(query=search_query, max_results=3)

context_from_web = ""
for idx, result in enumerate(search_result['results']):
    context_from_web += f"\nCorporate Competitor {idx+1}: {result['title']}\nURL: {result['url']}\nContent: {result['content']}\n"

# =====================================================================
# GENERATION: Corporate Comparison Analysis via OpenAI GPT-4o-mini
# =====================================================================
system_prompt = f"""
You are a senior Corporate Strategy Consultant working for "{company_name}".
Your output must strictly reflect the company's designated Brand Voice: "{brand_voice}".
"""

user_prompt = f"""
Your task is to conduct a corporate-level market research analysis comparing our firm "{company_name}" against the real-world competitor companies found in the web context.

Focus your analysis at the company-to-company level (compare overall market standing, company target audience, general pricing strategies, and firm-wide weaknesses) rather than just looking at isolated product features.

Our Company Details:
- Company Name: {company_name}
- Our Products & Pricing: {products_and_prices}

Live Web Corporate Competitor Context:
{context_from_web}

Expected Output:
A comprehensive markdown report listing the top 3 competitor companies. For each company include:
1. Overview of the Competitor Firm.
2. Estimated Corporate Pricing Strategy.
3. At least 2 major Corporate Vulnerabilities/Weaknesses compared to our firm.
All delivered strictly in our brand voice: "{brand_voice}".
"""

print(f"\n--- Kicking off OpenAI Corporate Research Agent for {company_name} ---")
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2
)

agent_final_output = response.choices[0].message.content

# Print the main market research output
print(f"\n================ SYSTEM RESEARCH OUTPUT FOR {company_name.upper()} ================")
print(agent_final_output)
```

```python
import os
import re
import json
from google.colab import userdata
from openai import OpenAI
from tavily import TavilyClient
from pypdf import PdfReader
from pydantic import BaseModel, Field

# 1. Fetch and set API keys securely from Colab Secrets
if "OPENAI_API_KEY" not in os.environ:
    os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

if "TAVILY_API_KEY" not in os.environ:
    os.environ["TAVILY_API_KEY"] = userdata.get('TAVILY_API_KEY')

# 2. Initialize OpenAI and Tavily clients
openai_client = OpenAI()
tavily_client = TavilyClient()

# =====================================================================
# PYDANTIC DATA MODELS FOR VALIDATION & STRUCTURED OUTPUTS
# =====================================================================
class CompanyProfile(BaseModel):
    """Defines structural and type constraints for incoming corporate profile data."""
    company_name: str = Field(..., min_length=2, max_length=100, description="The legal or market name of the firm.")
    products_and_prices: str = Field(..., min_length=10, description="Detailed text listing the solutions and pricing models.")
    brand_voice: str = Field(default="Professional", description="The communication tone alignment.")

# =====================================================================
# SECURITY & VALIDATION FUNCTIONS (Input Validation & Prompt Injection Guard)
# =====================================================================
def validate_input_text(text: str) -> bool:
    """Validates user input against basic prompt injection patterns and malicious intent."""
    if not text.strip():
        return False

    malicious_keywords = [
        "ignore previous instructions", "ignore all instructions", "override system",
        "system prompt", "you must forget", "act as a sudo", "bypass restrictions",
        "drop table", "select * from", "delete from"
    ]

    lowered_text = text.lower()
    for keyword in malicious_keywords:
        if keyword in lowered_text:
            print(f"\n[SECURITY ERROR]: Malicious keyword detected: '{keyword}'")
            return False

    if len(text) > 4000:
        print("\n[VALIDATION ERROR]: Input text size exceeds the allowed limit.")
        return False

    return True

def extract_text_from_pdf(pdf_path: str) -> str:
    """Safely reads and extracts text content from an uploaded PDF file."""
    try:
        reader = PdfReader(pdf_path)
        extracted_text = ""
        for page in reader.pages:
            text = page.extract_text()
            if text:
                extracted_text += text + "\n"
        return extracted_text
    except Exception as e:
        print(f"\n[VALIDATION ERROR]: Failed to process the PDF file. Error: {str(e)}")
        return ""

# =====================================================================
# DATA INGESTION & PARSING LAYER
# =====================================================================
print("================ COMPANY PROFILE ACQUISITION ================\n")
input_method = input("Do you want to enter data manually or upload a PDF? (type 'manual' or 'pdf'): ").strip().lower()

current_profile = None

if input_method == "pdf":
    pdf_file_name = input("Enter the PDF file name uploaded in Colab (e.g., profile.pdf): ").strip()
    if not os.path.exists(pdf_file_name):
        raise ValueError("[ERROR]: File not found. Please upload the PDF to the Colab sidebar first.")

    raw_pdf_content = extract_text_from_pdf(pdf_file_name)

    if not validate_input_text(raw_pdf_content):
        raise ValueError("[SECURITY EXCEPTION]: PDF content violated system security guidelines.")

    print("\nProcessing profile details via OpenAI Structured Outputs (Pydantic parsing)...")

    parse_response = openai_client.beta.chat.completions.parse(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Extract corporate profiling structures dynamically from unstructured raw documents."},
            {"role": "user", "content": f"Parse the following text content:\n{raw_pdf_content}"}
        ],
        response_format=CompanyProfile,
        temperature=0.1
    )
    current_profile = parse_response.choices[0].message.parsed

elif input_method == "manual":
    comp_name = input("Enter Company Name: ").strip()
    prod_price = input("Enter Products/Solutions Description with Pricing: ").strip()
    voice = input("Enter Brand Voice (e.g., Bold, Corporate, Friendly): ").strip()

    if not (validate_input_text(comp_name) and validate_input_text(prod_price) and validate_input_text(voice)):
        raise ValueError("[SECURITY EXCEPTION]: Input execution blocked due to invalid entries.")

    current_profile = CompanyProfile(company_name=comp_name, products_and_prices=prod_price, brand_voice=voice)
else:
    raise ValueError("[VALIDATION ERROR]: Invalid choice. You must pick 'manual' or 'pdf'.")

company_name = current_profile.company_name
products_and_prices = current_profile.products_and_prices
brand_voice = current_profile.brand_voice

# =====================================================================
# HUMAN IN THE LOOP (HITL) GATEWAY
# =====================================================================
print("\n================ HUMAN-IN-THE-LOOP VERIFICATION ================")
print(f"Company Name: {company_name}")
print(f"Products & Pricing Structure: {products_and_prices}")
print(f"Brand Voice Alignment: {brand_voice}")
print("=================================================================")

human_approval = input("Review the validated profile data above. Do you approve it? (type 'yes' to proceed, 'no' to abort): ").strip().lower()

if human_approval != 'yes':
    print("\n[SYSTEM TERMINATION]: Execution stopped by user during the validation loop.")
    raise SystemExit

print("\n--- Human Verification Passed. Proceeding to execution phase. ---")

# =====================================================================
# RETRIEVAL: Corporate Intelligence Acquisition via Tavily API (UPDATED LIMIT)
# =====================================================================
# Increased cut-off limit: Trim products description to 250 characters for richer context
clean_products_summary = products_and_prices[:250] if len(products_and_prices) > 250 else products_and_prices

search_query = f"top competitor companies competing with {company_name} in {clean_products_summary}"

# Safety net fallback to guarantee keeping total characters below Tavily's 400 boundary
if len(search_query) > 380:
    search_query = f"top competitor firms competing with {company_name}"

print(f"\n--- Tavily is searching live web using safe query ({len(search_query)} chars): '{search_query}' ---")
search_result = tavily_client.search(query=search_query, max_results=3)

context_from_web = ""
for idx, result in enumerate(search_result['results']):
    context_from_web += f"\nCorporate Competitor {idx+1}: {result['title']}\nURL: {result['url']}\nContent: {result['content']}\n"

# =====================================================================
# GENERATION: Corporate Comparison Analysis via OpenAI GPT-4o-mini
# =====================================================================
system_prompt = f"""
You are a senior Corporate Strategy Consultant working for "{company_name}".
Your output must strictly reflect the company's designated Brand Voice: "{brand_voice}".
"""

user_prompt = f"""
Your task is to conduct a corporate-level market research analysis comparing our firm "{company_name}" against the real-world competitor companies found in the web context.

Focus your analysis at the company-to-company level (compare overall market standing, company target audience, general pricing strategies, and firm-wide weaknesses) rather than just looking at isolated product features.

Our Company Details:
- Company Name: {company_name}
- Our Products & Pricing: {products_and_prices}

Live Web Corporate Competitor Context:
{context_from_web}

Expected Output:
A comprehensive markdown report listing the top 3 competitor companies. For each company include:
1. Overview of the Competitor Firm.
2. Estimated Corporate Pricing Strategy.
3. At least 2 major Corporate Vulnerabilities/Weaknesses compared to our firm.
All delivered strictly in our brand voice: "{brand_voice}".
"""

print(f"\n--- Kicking off OpenAI Corporate Research Agent for {company_name} ---")
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2
)

agent_final_output = response.choices[0].message.content

# Print the main market research output
print(f"\n================ SYSTEM RESEARCH OUTPUT FOR {company_name.upper()} ================")
print(agent_final_output)
```

#### EVALUATION LAYER: LLM-as-a-Judge Validation via Pydantic

```python
# =====================================================================
# EVALUATION LAYER: LLM-as-a-Judge Validation via Pydantic
# =====================================================================
class RAGEvaluationSchema(BaseModel):
    """Defines strict validation rules and ranges for RAG quality assurance scoring."""
    groundedness_score: int = Field(..., ge=1, le=5, description="Score 1-5 verifying absence of hallucinations.")
    groundedness_reason: str = Field(..., description="Justification text for the groundedness score.")
    context_relevance_score: int = Field(..., ge=1, le=5, description="Score 1-5 verifying necessity of data retrieval.")
    context_relevance_reason: str = Field(..., description="Justification text for context relevance.")
    answer_relevance_score: int = Field(..., ge=1, le=5, description="Score 1-5 verifying fulfillment of explicit prompts.")
    answer_relevance_reason: str = Field(..., description="Justification text for answer relevance.")

evaluation_user_prompt = f"""
Evaluate the following RAG interaction:

[User Corporate Input]: Company: {company_name} | Pricing & Products: {products_and_prices}
[Retrieved Web Context]: {context_from_web}
[Agent Final Output]: {agent_final_output}
"""

print(f"\n--- Running Pydantic Structured RAG Evaluation ---")

eval_response = openai_client.beta.chat.completions.parse(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are an expert AI Quality Assurance Judge specialized in analyzing RAG systems for commercial agents."},
        {"role": "user", "content": evaluation_user_prompt}
    ],
    response_format=RAGEvaluationSchema,
    temperature=0.1
)

eval_results = eval_response.choices[0].message.parsed

# Display the final structured evaluations safely extracted via Pydantic properties
print(f"\n================ RAG EVALUATION REPORT ================")
print(f"Groundedness Score: {eval_results.groundedness_score}/5")
print(f"Reason: {eval_results.groundedness_reason}\n")

print(f"Context Relevance Score: {eval_results.context_relevance_score}/5")
print(f"Reason: {eval_results.context_relevance_reason}\n")

print(f"Answer Relevance Score: {eval_results.answer_relevance_score}/5")
print(f"Reason: {eval_results.answer_relevance_reason}")
print(f"=======================================================")
```
